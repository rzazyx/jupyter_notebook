```python
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x=np.linspace(-5, 5, 100)
y=np.linspace(-5, 5, 100)
X,Y=np.meshgrid(x,y)
def f(x,y):
  return np.sin(np.sqrt(x**2 + y**2))
Z=f(X,Y)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
contour = ax.contour3D(X, Y, Z, 50, cmap='viridis')

#warna
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
fig.colorbar(contour, ax=ax, label='Z values')
plt.show()
```

```python
import os
import zipfile
from PyPDF2 import PdfReader
from pdf2image import convert_from_path
from google.colab import files
from IPython.display import display, Image

def pdf_to_images(pdf_file, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    images = convert_from_path(pdf_file, poppler_path='/usr/bin')
    image_paths = []
    for i, image in enumerate(images):
        img_path = os.path.join(output_dir, f"page_{i}.png")
        image.save(img_path, 'PNG')
        image_paths.append(img_path)
    return image_paths

def create_zip(file_paths, zip_name):
    with zipfile.ZipFile(zip_name, 'w') as zipf:
        for file in file_paths:
            zipf.write(file, os.path.basename(file))

output_dir = "output_images"
zip_name = "images.zip"

# Konversi PDF ke gambar
image_paths = pdf_to_images(pdf_file, output_dir)

# Tampilkan gambar-gambar hasil konversi
for img_path in image_paths:
    display(Image(filename=img_path))

# Buat file ZIP dari gambar-gambar tersebut
create_zip(image_paths, zip_name)

# Unduh file ZIP
files.download(zip_name)
```
